```python
#Create 3 DataFrames and save them as CSV files
import pandas as pd

project_data = {
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5",
                "Project 6","Project 7","Project 8","Project 9","Project 10",
                "Project 11","Project 12","Project 13","Project 14"],
    "Cost": [1002000,2000000,4500000,5500000,None,680000,400000,350000,None,300000,2000000,1000000,3000000,200000],
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished",
               "Failed","Finished","Failed","Ongoing","Finished",
               "Failed","Ongoing","Finished","Finished"]
}

project_df = pd.DataFrame(project_data)
project_df.to_csv("project.csv", index=False)


# ---------------- Employee DataFrame ----------------
employee_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Name": ["John Alter","Alice Luxemburg","Tom Sebastine","Nina Adgra","Amy Johny"],
    "Gender": ["M","F","M","F","F"],
    "City": ["Paris","London","Berlin","Newyork","Madrid"],
    "Age": [25,27,29,31,30]
}

employee_df = pd.DataFrame(employee_data)
employee_df.to_csv("employee.csv", index=False)


# ---------------- Seniority Level DataFrame ----------------
seniority_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Designation Level": [2,2,3,2,3]
}

seniority_df = pd.DataFrame(seniority_data)
seniority_df.to_csv("seniority.csv", index=False)
```

```python
#Fill missing Cost values using average (must use for loop)
avg=project_df["Cost"].mean()
for i in range(len(project_df)):
    if pd.isna(project_df.loc[i,'Cost']):
        project_df.loc[i,'Cost']=avg
print(project_df)
```

```python
#Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column.
first_names = []
last_names = []

for name in employee_df["Name"]:
    # Split the name using space
    parts = name.split(" ")
    # Append first part as first name
    first_names.append(parts[0])
    # Append second part as last name
    last_names.append(parts[1])
# Create new columns
employee_df["First Name"] = first_names
employee_df["Last Name"] = last_names

# Drop the old Name column
employee_df.drop(columns=["Name"], inplace=True)
print(employee_df)
```

```python
print(employee_df.loc[:, ['Age', 'City']])
```

```python
# Join all three dataframes in one single dataframe. Name it “Final” 
# Merge employee and seniority data
final_df = pd.merge(employee_df, seniority_df, on="ID", how="inner")

# Merge project data
final_df = pd.merge(final_df, project_df, on="ID", how="inner")
```

```python
print(final_df)
```

```python
#Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects. 
final_df['Bonus']=0
for i in range(len(final_df)):
    if final_df.loc[i, "Status"] == "Finished":
        final_df.loc[i, "Bonus"] = final_df.loc[i, "Cost"] * 0.05
print(final_df)
```

```python
#Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.
# Demote designation for failed projects
for i in range(len(final_df)):
    if final_df.loc[i, "Status"] == "Failed":
        final_df.loc[i, "Designation Level"] += 1

# Remove employees whose designation level > 4
final_df = final_df[final_df["Designation Level"] <= 4]
```

```python
print(final_df)
```

```python
#Add “Mr.” and “Mrs.” to the first name column and drop the gender column. 
# Add title based on gender
for i in range(len(final_df)):
    if final_df.loc[i, "Gender"] == "M":
        final_df.loc[i, "First Name"] = "Mr. " + final_df.loc[i, "First Name"]
    else:
        final_df.loc[i, "First Name"] = "Mrs. " + final_df.loc[i, "First Name"]

# Drop gender column
final_df.drop(columns=["Gender"], inplace=True)
```

```python
print(final_df)
```

```python
#Promote designation level by 1 for the employees whose age is more than 29 years using IF condition. 
# Promote based on age
for i in range(len(final_df)):
    if final_df.loc[i, "Age"] > 29:
        final_df.loc[i, "Designation Level"] -= 1
print(final_df)
```

```python
#Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost
# Group by employee and sum cost
total_cost_df = final_df.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Rename column
total_cost_df.rename(columns={"Cost": "TotalProjCost"}, inplace=True)
print(total_cost_df)
```
